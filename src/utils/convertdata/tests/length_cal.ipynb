```python
import numpy as np

OFFSET_FROM_ZERO = 29 # mm
MM_SCALE =  np.arange(0,3000,100) + OFFSET_FROM_ZERO
MM_SCALE =  np.arange(0,3000,100) 
V_SCALE = np.array([
    211, 1083, 1959, 2821, 3704,
    4585, 5498, 6369, 7243, 8132, 
    9034, 9906, 10798, 11691, 12596,
    13469, 14366, 15248, 16140, 17016, 
    17909, 18805, 19707, 20589, 21487, 
    22361, 23254, 24132, 25033, 25924
])

LENGTH_OFFSET = 35-2+200
```

```python
res = np.polyfit(V_SCALE, MM_SCALE, 1)
p = np.poly1d(res)
```

```python
def calc_length(length: float) -> float:
    return p(length) 
    return p(length) + LENGTH_OFFSET
```

```python
calc_length(1104)
```

```python
calc_length(3735) - calc_length(3565)
```

```python
def calc_length_simple(length):
    return (length-V_SCALE[0])*(3000/V_SCALE[-1])
```

```python
calc_length_simple(1999)
```
